# ML

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['Ship Mode'] = le.fit_transform(df['Ship Mode'])
df['Segment'] = le.fit_transform(df['Segment'])
df['Category'] = le.fit_transform(df['Category'])
df['Sub-Category'] = le.fit_transform(df['Sub-Category'])
df['Order Priority'] = le.fit_transform(df['Order Priority'])
df.head()
```

```python
# Splitting the data into train and test sets
X = df[['Ship Mode','Segment','Category','Sub-Category','Sales','Quantity','Discount','Shipping Cost','Order Priority']]
Y = df['Profit']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
X.info()
```

```python
model = RandomForestRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
pred = model.predict(X_test)
```

```python
mse = mean_squared_error(Y_test, pred)
print(f'Mean Squared Error: {mse}')
```

```python
from sklearn.metrics import r2_score, mean_absolute_error

# Calculate additional metrics
r2 = r2_score(Y_test, pred)
mae = mean_absolute_error(Y_test, pred)
rmse = mean_squared_error(Y_test, pred, squared=False)

print(f'R-squared: {r2}')
print(f'Mean Absolute Error: {mae}')
print(f'Root Mean Squared Error: {rmse}')
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=Y_test, y=pred)
plt.xlabel('Actual Profit')
plt.ylabel('Predicted Profit')
plt.title('Actual vs Predicted Profit')
plt.show()
```

The provided scatter plot shows the relationship between the actual profit values (on the x-axis) and the predicted profit values (on the y-axis) from the RandomForestRegressor model. Here are some insights based on the scatter plot:

Positive Correlation:

There is a noticeable positive correlation between the actual and predicted profit values. This indicates that as the actual profit increases, the predicted profit also tends to increase, suggesting that the model has learned some relationship between the features and the target variable.
Clustered Predictions:

Most of the data points are clustered around the line where the actual profit equals the predicted profit (the line y = x). This indicates that for many observations, the predicted profit is close to the actual profit, which is a good sign of model performance.
Outliers:

There are a few outliers where the predicted profit significantly deviates from the actual profit. These outliers could be due to various reasons, such as:
The model may not have enough information to accurately predict these specific cases.
There might be anomalies or noise in the data.
Certain features that influence profit significantly may be missing from the model.
Model Performance:

Although there are outliers, the overall trend suggests that the model performs reasonably well for a majority of the data points. To quantitatively evaluate the model performance, we should look at metrics like Mean Squared Error (MSE), Mean Absolute Error (MAE), and RÂ² score, as calculated previously.
